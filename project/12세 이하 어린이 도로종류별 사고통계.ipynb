### 보행자 무단횡단 사고

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\User\\Desktop\\데마\\팀프\\mudanwalking.csv", encoding='CP949')
```

```python
df
```

```python
df.describe()
```

```python
df2 = df[['발생건수', '경도', '위도']]
df2 = df2.sort_values(by='발생건수').reset_index()
```

```python
df2_bot = df2.head(100)
df2_top = df2.tail(100)
```

```python
import folium
map_osm = folium.Map(location=[37.507161, 126.951753], zoom_start=12)
```

```python
for item in df2_top.index:
    latitude = df2_top.loc[item, '위도']
    longitude = df2_top.loc[item, '경도']
    
    folium.CircleMarker([latitude, longitude],
                       radius=df2_top.loc[item, '발생건수']/1,
                       color = 'blue',
                       fill = True).add_to(map_osm)
for item in df2_bot.index:
    latitude = df2_bot.loc[item, '위도']
    longitude = df2_bot.loc[item, '경도']
    
    folium.CircleMarker([latitude, longitude],
                       radius=df2_bot.loc[item, '발생건수']/1,
                       color = 'red',
                       fill = True).add_to(map_osm)
```

```python
map_osm
```
