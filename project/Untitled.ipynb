```python
import folium
map_osm = folium.Map(location=[37.507161, 126.951753], zoom_start=12)
```

```python
df_sub = df[['경도', '위도', '사상자수', '사망자수']]
df2 = df_sub.sort_values(by='사망자수').head(100).reset_index()
df3 = df_sub.sort_values(by='사상자수').head(100).reset_index()
```

```python
for item in df2.index:
    latitude = df2.loc[item, '위도']
    longitude = df2.loc[item, '경도']
    
    folium.CircleMarker([latitude, longitude],
                       radius=df2.loc[item, '사망자수']/1,
                       color = 'blue',
                       fill = True).add_to(map_osm)
for item in df3.index:
    latitude = df3.loc[item, '위도']
    longitude = df3.loc[item, '경도']
    
    folium.CircleMarker([latitude, longitude],
                       radius=df3.loc[item, '사상자수']/1,
                       color = 'red',
                       fill = True).add_to(map_osm)
```

```python
from folium import plugins

location=df['위도'].mean(), df['경도'].mean()

#A basemap is then created using the location to centre on and the zoom level to start.
m = folium.Map(location=location,zoom_start=15)

#Each location in the DataFrame is then added as a marker to the basemap points are then added to the map
for i in range(0,len(df)):
     num_of_casualties = df['사망자수'].iloc[i]
     if num_of_casualties == 1:
        color = 'blue'
     elif num_of_casualties == 2:
        color = 'green'
     else:
        color = 'red'
        
     #popup = folium.Popup('Accident', parse_html=True) 
     folium.Marker([df['위도'].iloc[i],df['경도'].iloc[i]], icon=folium.Icon(color=color, icon='info-sign')).add_to(m)
        
m
```
