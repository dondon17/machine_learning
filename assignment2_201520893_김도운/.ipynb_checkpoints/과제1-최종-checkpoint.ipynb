```python
from sklearn.datasets import fetch_california_housing

california = fetch_california_housing()
```

```python
import pandas as pd
df = pd.DataFrame(california.data, columns=california.feature_names)
df_target = pd.DataFrame(california.target, columns=["Target"])
```

```python
import numpy as np
```

```python
bins = [0.2, 1.7, 2.5, 5]
labels = ["level_1", "level_2", "level_3"]
```

```python
def make_onehot(df, target, bins, labels):
    target = pd.cut(target['Target'], bins= bins, labels= labels)
    target = target.dropna()
    target = pd.get_dummies(target)
    
    tpp = pd.concat([df, target], axis= 1, join= "inner")
    return tpp
```

```python
data = make_onehot(df, target = df_target, bins= bins, labels=labels)
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,4))
plt.xlabel('Medlnc', fontsize= 14)
plt.ylabel('Longitude', fontsize= 14)
plt.title('problem 2')
plt.scatter(data.MedInc, data.Longitude, s= data.level_1, c= 'r', label= labels[0])
plt.scatter(data.MedInc, data.Longitude, s= data.level_2, c= 'b', label= labels[1])
plt.scatter(data.MedInc, data.Longitude, s= data.level_3, c= 'purple', label= labels[2])
plt.legend()
plt.show()
```
