## US Cars Dataset

```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\User\\Desktop\\데마\\중간고사 대체과제\\dataset\\USA_cars.csv")
df.head()
```

```python
df.drop(['Unnamed: 0', 'vin', 'lot', 'country'], axis=1, inplace=True)
df.dropna()
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
sns.scatterplot(x=df['price'], y=df['year'], hue=df['brand']
               , s= 100)
plt.xlabel("price", size=20)
plt.ylabel("year", size=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
df['state'].value_counts().plot(kind='barh')
plt.show()
# sns.catplot(x=df['state'].sort_values, y=df['price'], data=df)
```
