## Airline Passenger Satisfaction

```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv("C:\\Users\\User\\Desktop\\데마\\중간고사 대체과제\\dataset\\Airline_satisfaction.csv", encoding='utf-8')
df_target = df['satisfaction']
df.head(10)
```

```python
def satisfaction_to_num(x):
    if x == 'satisfied': return 1
    elif x == 'neutral or dissatisfied': return 0
    else: return -1
    
df['satisfaction'] = df['satisfaction'].apply(satisfaction_to_num)
df.drop(["Unnamed: 0", "id"], inplace=True, axis=1)
df.dropna()
df.head(10)
```

```python
# 좌석 클래스, 여행 유형에 따른 만족도 분포
plt.figure(figsize=(20, 20))

plt.subplot(221)
sns.boxplot(x='Class', y='satisfaction', data=df)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.subplot(222)
sns.boxplot(x='Type of Travel', y='satisfaction', data=df)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()
```

```python
gen2num = pd.get_dummies(df['Gender'],drop_first=False)
cus2num = pd.get_dummies(df['Customer Type'],drop_first=False)
type2num = pd.get_dummies(df['Type of Travel'],drop_first=False)
class2num = pd.get_dummies(df['Class'],drop_first=False)
df = pd.concat([df,gen2num,cus2num,type2num,class2num],axis =1)
df.drop(['Gender','Customer Type','Type of Travel','Class'],inplace =True,axis = 1)
df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(25,21))
mask = np.zeros_like(df.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(df.corr(),annot=True,cmap='binary', mask=mask)
plt.show()
```

```python
dataCorr = df.corr()
dataCorr = dataCorr[abs(dataCorr) >= 0.01].stack().reset_index()
dataCorr = dataCorr[dataCorr['level_0'].astype(str)!=dataCorr['level_1'].astype(str)]

dataCorr['ordered-cols'] = dataCorr.apply(lambda x: '-'.join(sorted([x['level_0'],x['level_1']])),axis=1)
dataCorr = dataCorr.drop_duplicates(['ordered-cols'])
dataCorr.drop(['ordered-cols'], axis=1, inplace=True)
 
dataCorr.sort_values(by=[0], ascending=False).head(5)
```
